### Data Preprocessing
Step 1: Converting object columns to numeric

```python
# Even after replacing the Nan in ca and Thal the datatype shows object

df['ca'] = pd.to_numeric(df['ca'])
df['thal'] = pd.to_numeric(df['thal'])
```

```python
df.info()
```

### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X = df.drop('target', axis=1)
y = df['target']

X_scaled = scaler.fit_transform(X)
```

```python
print(X_scaled[:5])
```

### Applying Train & Test Split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

print("Training size:", X_train.shape)
print("Training size:", X_test.shape)
```

### Step 3: Model Building — Logistic Regression


```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)

print("Model Training Complete")
```

```python
y_pred = model.predict(X_test)
print(y_pred)
```

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test,y_pred)
print("Model Accurcacy:", accuracy*100, "%")
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```
